```python
from DNAFlex.psf import *
from DNAFlex.stats import PoissonNormal, PoissonNormalApproximate
from scipy.stats import multivariate_normal
import numpy as np
```

```python
mu_norm = 100 #ADU
mu_psn = 1000 #e-
sigma_norm = 0.1 #ADU
gain = 2.2 #ADU/e-
pnorm = PoissonNormal(mu_norm,mu_psn,sigma_norm)
pnorm_approx = PoissonNormalApproximate(mu_norm,mu_psn,sigma_norm)
```

```python
x = np.arange(0,5000,1)
pmf = pnorm.get_pmf(x)
pmf_approx = pnorm_approx.get_pmf(x)
```

```python
samples = pnorm.sample(1000)
plt.hist(samples,density=True,color='gray')
plt.plot(x,pmf,color='red')
plt.plot(x+mu_norm-sigma_norm**2,pmf_approx,color='blue',linestyle='--')
plt.xlim([900,1300])
plt.show()
```
