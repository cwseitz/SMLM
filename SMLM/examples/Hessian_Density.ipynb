```python
from SMLM.psf import *
from scipy.stats import multivariate_normal
import numpy as np
```

```python
#########################
# Parameters
#########################

L = 100
omat = np.ones((L,L))
gain0 = 2.2 #huang 2013
offset0 = 0.0
var0 = 100.0
gain = gain0*omat #ADU/e-
offset = offset0*omat #ADU
var = var0*omat #ADU^2
pixel_size = 108.3 #nm
sigma = 0.22*640/1.4 #zhang 2007
sigma = sigma = sigma/pixel_size
texp = 1.0 #seconds
eta = 0.8 #quantum efficiency
N0 = 1000
B0 = 0
```

```python
nspots = 100
theta = np.zeros((5,nspots))
theta[0,:] = np.random.uniform(10,L-10,nspots)
theta[1,:] = np.random.uniform(10,L-10,nspots)
theta[2,:] = sigma
theta[3,:] = N0
theta[4,:] = B0
```

```python
frame = Frame(theta,eta,texp,L,gain,offset,var)
frame.generate(plot=True)
```

```python
errs = get_errors(theta,frame.adu,eta,texp,gain,var,plot=True)
```
